```python
#Задание 1 
str_1 = 'AaB'
str_2 = 'Ab'
is_the_same_letters = set(str_1.lower()) == set(str_2.lower())

print("Результат :", is_the_same_letters)
```

```python
#Задание 2
x = '3.14'
y = int(x.split('.')[1])

print("Результат :", y)
```

```python
#Задание 3
str_1 = "Кот"
str_2 = "Ток"
result = sorted(str_1.lower()) == sorted(str_2.lower())

print("Результат :", result)
```

```python
#Задание 4
a = "Я     изучаю   новый язык      программирования.   "
result = " ".join(a.split())
print("Результат :", result)
```

```python
#Задание 5 
a = "li*3"
tag, count = a.split('*')
count = int(count)
result = "".join([f"<{tag}></{tag}>" for _ in range(count)])

print("Результат :", result)
```

```python
#Задание 6
x = 'У лукоморья дуб зелёный'
result = " ".join(sorted(x.split()))

print("Результат :", result)
```

```python
#Задание 7 
x = 'На подоконнике в гостиной у бабушки росли цветы, цветы были красные и желтые, очень красивые цветы.'
y = 'цветы'
result = x.count(y)

print("Результат :", result)
```

```python
#Задание 8
card = '5468350018455833'
result = '*' * (len(card) - 4) + card[-4:]

print("Результат :", result)
```
