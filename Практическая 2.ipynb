```python
#Задание 6
import random

numbers = []
for i in range(10):
    numbers.append(random.randint(10, 100))

total = sum(numbers)
average = total / len(numbers)

print(numbers)
print("Сумма:", total)
print("Среднее значение:", round(average, 2))
```

```python
#Задание 7 
lst = [1, 2, 3, 4, 5, 5, 4, 3, 2, 1, 5]
lst.insert(lst.index(2) + 1, 'a')

numbers = [random.randint(10, 100) for _ in range(5)]
lst += numbers

count_fives = lst.count(5)
print(lst)
print("Колличество чисел 5:", count_fives)
```

```python
#Задание 8 
text = input("Введите текст: ")

words = text.split()
average_lenght = sum(len(word.strip('.,!?'))for word in words) / len(words)

print("Средняя длина слова:", round(average_lenght, 2))
```
