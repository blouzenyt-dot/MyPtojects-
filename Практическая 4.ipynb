```python
#Задание 2 
data = {
    "cost_price": 225.89,
    "sell_price": 550.00,
    "eclairs": 100
}
profit = (data["sell_price"] - data["cost_price"]) * data["eclairs"]
result = round(profit)

print("Результат -", result)
```
