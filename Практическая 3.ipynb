```python
#Задание 1 
numbers_list = [1, 5, 3, 3, 5]
numbers_list_ordered = sorted(numbers_list, reverse=True)
numbers_set = set(numbers_list)
numbers_set.add(max(numbers_list) + 1)

unique_numbers = set(numbers_list)
unique_numbers.discard(min(unique_numbers))
numbers_frozenset = frozenset(list(unique_numbers))

print("numbers_list_ordered =", numbers_list_ordered)
print("numbers_set = set(", list(numbers_set), ")", sep="")
print("numbers_frozenset = frozenset(", list(numbers_frozenset), ")", sep="")
print()
```

```python
#Задача 2
list_1 = [1, 5, 3]
list_2 = [2, 8]

list_1.sort()
list_2.sort(reverse = True)
list_3 = sorted(list_1 + list_2)
list_3_len = len(list_3)

print("list_1 =", list_1)
print("list_2 =", list_2)
print("list_3 =", list_3)
print("list_3_len =", list_3_len)
```

```python
#Задача 4 
my_set = {0, 10, 100}
to_delete = 0

if to_delete in my_set:
    my_set.discard(to_delete)
    print(f"Число {to_delete} удалено из множества.")
else:
    print(f"Числа {to_delete} нет в множестве.")

print("my_set =", list(my_set))
```

```python
#Задача 5 
students = {'Кравченко Виталий', 'Полякова Ольга', 'Некрасов Игорь',
            'Дудочкин Илья', 'Захарова Мария'}
new_student = 'Василенко Анна'
churn_student = 'Полякова Ольга'

students.add(new_student)
students.discard(churn_student)

students_sorted = sorted(students)

print("students =", students_sorted)
```

```python
#Задача 6 
da_students = {'Ivanov Alexander', 'Loginov Vladislav', 'Ershova Anna', 'Korneva Daria'}
dv_students = {'Ershova Anna', 'Egunov Andrey', 'Ignatov Alexey', 'Loginov Vladislav'}

studenst_da_dv = da_students.intersection(dv_students)
studenst_da_dv_sorted = sorted(studenst_da_dv)

print("students =", studenst_da_dv_sorted)
```

```python
#Задача 7
a = [1, 2, 4, 3]

mid = len(a) // 2
sum_left = sum(a[:mid])
sum_right = sum(a[mid:])
result = sum_left == sum_right 

print("sum_left =", sum_left)
print("sum_right =", sum_right)
print("result =", result)
```
